```python
import pandas as pd
import numpy as np
import geopandas as gdp
```

```python
peru_shapefile = gdp.read_file('anexos\distritos\distritos-peru@bogota-laburbano.geojson')

peru_shapefile
```

```python
import matplotlib.pyplot as plt

peru_shapefile = peru_shapefile.loc[(peru_shapefile['nombdep'] == 'LIMA'),:]

lima_metropolitana = peru_shapefile[(peru_shapefile['nombdep'] == 'LIMA') & 
                              (peru_shapefile['nombprov'] == 'LIMA')]

callao = peru_shapefile[peru_shapefile['nombdep'] == 'CALLAO']

# Plotting
fig, ax = plt.subplots(figsize=(10, 10))
peru_shapefile.plot(ax=ax, color='lightgrey', edgecolor='black')
lima_metropolitana.plot(ax=ax, edgecolor='blue', linewidth=2, facecolor='none')
callao.plot(ax=ax, edgecolor='blue', linewidth=2, facecolor='none')

ax.set_title('Distritos de Lima Metropolitana y Callao')
ax.set_axis_off()

plt.show()
```
